```python
import numpy as np
from numpy import arange, hamming, sin, pi
from matplotlib import pylab as plt
```

```python
N=128
dt=0.01
freq=10
amp=1
t=np.arange(0,N*dt,dt)
f=amp*np.sin(2*np.pi*freq*t)

plt.xlabel('time(sec)', fontsize=14)
plt.ylabel('signal amplitude', fontsize=14)
plt.plot(t,f)
```

```python
F = np.fft.fft(f)
```

```python
F_abs=abs(F)
F_abs_amp=F_abs/N*2
F_abs_amp[0] /= 2
plt.plot(F_abs_amp)
```

```python
fq = np.linspace(0, 1.0/dt, N)
```

```python
plt.xlabel('freqency(Hz)', fontsize=14)
plt.ylabel('signal amplitude', fontsize=14)
plt.plot(fq[:N//2+1], F_abs_amp[:N//2+1]) # 
```
